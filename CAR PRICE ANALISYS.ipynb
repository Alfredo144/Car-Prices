CORRELACIÓN PEARSON

```python
from scipy import stats
```

```python
#Calcularé el valor P y el coeficiente de correlación entre el precio del carro y la potencia del motor
pearson_coef, p_value=stats.pearsonr(df["horsepower"],df["price"])
pearson_coef, p_value
```

```python
#y también entre el precio y el tamaño del motor
pearson_coef1, p_value1=stats.pearsonr(df["engine-size"],df["price"])
pearson_coef1, p_value1
```

```python
#y para el ancho del auto y su precio:
pearson_coef2, p_value2=stats.pearsonr(df["bore"],df["price"])
pearson_coef2, p_value2
```

Se puede observar en los dos primeros un coeficiente de correlación positivo y muy cercano a 1. Esto indica que las variables en cuestión son buenos predictores de precio. Además cuentan con valores P muy pequeños. 

Para el tercer caso, no parece haber una correlación significativa entre el diámetro del pistón y el precio del auto, por lo tanto, debe considerarse descartar esta variable si se busca predecir el precio.

Ahora, buscando una correlación entre las 3 categorías del tipo de tracción, y el precio promedio de cada uno, es conveniente hacer un ANOVA.


```python
#Primero agrupo todos los elementos de cada categoría de la variable "drive-wheels" y el precio de cada una
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
df_gptest
```

```python
#y ahora calculo los valor p y f-test
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

Este resultado -F grande y P pequeño- muestran una gran correlación y gran significancia estadística. ¿Esto significa que los 3 grupos están altamente correlacionados?

```python
# veamos con fwd y rwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```python
#ahora con 4wd y rwd 
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
# y finalmente con 4wd y fwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])  
 
print("ANOVA results: F=", f_val, ", P =", p_val)   
```
